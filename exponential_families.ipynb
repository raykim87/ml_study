```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def normal_curve(x, mu, std, c):
    return(1/(np.pi * 2*std) * (c) * np.exp(-(x-mu)**2/(2*std)))
```

```python
xs = np.arange(-10, 10, 0.1)
```

```python
plt.plot(normal_curve(xs, 0, 1, 5))
plt.plot(normal_curve(xs, 5, 1, 5**(1/2)))
```

```python
normal_curve(0,0,1)
```

```python
def expoential_dist(x, theta):
    ys=[]
    for val in x:
        if val>0: ys.append(theta * math.e**(-theta * val))
        else: ys.append(0) 
    return(ys)
```

```python
plt.plot(xs, expoential_dist(xs, theta=0.1))
plt.plot(xs, expoential_dist(xs, theta=0.5))
```
